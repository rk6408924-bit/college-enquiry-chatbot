````python
# 🎓 College Enquiry Chatbot (AI Project)

## Objective
To build a chatbot that answers student queries about courses, fees, admission, placements, and campus facilities using structured data.

## Technologies Used
- Python
- Jupyter Notebook
- Streamlit (GUI)
- JSON (Dataset)
````

````python
import json

# Load dataset
with open("../data/college_data.json") as f:
    data = json.load(f)

# Display data
data
````

````python
## Data Description

The dataset contains:
- College name
- Courses and fees
- Admission process
- Hostel details
- Facilities (Gym, Playground, etc.)
- Placements and internships
- Scholarships
- Campus life information
````

````python
query = "What are the fees for B.Tech?"
query.lower()
````

````python
def test_chatbot(query):
    query = query.lower()

    if "fee" in query:
        return data["courses"]

    elif "hostel" in query:
        return data["hostel_fees"]

    elif "facility" in query or "gym" in query or "playground" in query:
        return data["facilities"]

    elif "placement" in query:
        return data["placements"]

    elif "internship" in query:
        return data["internships"]

    elif "scholarship" in query:
        return data["scholarships"]

    elif "campus" in query or "vibe" in query:
        return data["campus_vibe"]

    else:
        return "Other query"

# Test single query
test_chatbot(query)
````

````python
## Conclusion

The chatbot successfully retrieves relevant information from the dataset based on user queries.

This logic is later integrated into a Streamlit-based GUI to create a user-friendly chatbot interface.

Future improvements:
- Integration with real AI models (LLM)
- Database storage
- Voice-based chatbot
````
